```python
import numpy as np
import pandas as pd
```

```python
items = pd.read_csv('/content/items.csv',encoding='latin1')
users = pd.read_csv('/content/Users.csv')
ratings = pd.read_csv('/content/Ratings.csv')
```

```python
#### popularity based recommender system 
```

```python
ratings_with_name = ratings.merge(items,on ='ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Product-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Product-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Product-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=5].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(items,on='Product-Title').drop_duplicates('Product-Title')[['Product-Title','Product-Company','Image-URL-S','num_ratings','avg_rating']]
```

```python
popular_df
```

```python
####Collaborative Filtering Based Recommender System
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 2
padhe_likhe_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y = filtered_rating.groupby('Product-Title').count()['Book-Rating']>=2
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Product-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Product-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = items[items['Product-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Product-Title')['Product-Title'].values))
       # item.extend(list(temp_df.drop_duplicates('Product-Title')['Company-Name'].values))
        item.extend(list(temp_df.drop_duplicates('Product-Title')['Image-URL-S'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('100% Compostable Paper Plates 10 Inch Bulk [500 Count] Heavy-Duty Dinner Plates with Compartments - Natural Brown Color Bagasse - Eco-Friendly Disposable Biodegradable Sugarcane - [Case 4/125 Pack]')
```
